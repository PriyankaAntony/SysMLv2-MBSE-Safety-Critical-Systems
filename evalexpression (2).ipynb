```sysml
package BatteryPowerCalculation {
    import ISQ::*;
    import SI::*;
    import Time::*;
    import ScalarValues::Real;

    // Energy Storage Requirement Calculation
    calc def RequiredBatteryCapacity {
        in nightHours : Real;
        in avgPowerDraw : PowerValue;
        
        return capacity : EnergyValue = avgPowerDraw * nightHours[h];
    }

   
    part def BatteryExample {
        // Assign values to attributes
        attribute nightHrs : Real = 12;          // 12 hours of darkness
        attribute avgDraw : PowerValue = 150[W]; // 150W average power draw
       
        // Bind calculation inputs
         capacity : EnergyValue=RequiredBatteryCapacity(nightHrs,avgDraw);
       
       
        }
    }

```
