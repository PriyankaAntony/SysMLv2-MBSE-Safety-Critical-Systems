```sysml
package HAPlogicalsub {

    part def Data;
    part def Energy;
    part def Signal;
    part def HAP;
    part def Image;
 
    port def Energyout {
        out item energyOut: Energy;
        in item energyIn: Energy;
    }
    port def EnergyIn {
        
        in item energyIn: Energy;
        out item energyOut: Energy;
    }
    
    port def SignalOut {
        out item signalOut: Signal;
        in item signalIn: Signal;
    }
     port def SignalIn {
         in item signalIn: Signal;
        out item signalOut: Signal;
        
    }
    port def ImageOut {
        
         out item imageOut: Image;
         in item imageIn: Image;
        
        
    }

    port def ImageIn {
         in item imageIn: Image;
        out item imageOut: Image;
        
    }

}
```
