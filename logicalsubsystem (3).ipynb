```sysml

package 'HAPLogicalSystemstructure' {
     import ISQ::*;
     import ScalarValues::*;
     import ScalarValues::Integer;
  part HAP{
      
  
   part  airframeStructure {
       attribute weight;
        
        part powersystem{
           attribute Energy;
           perform action StoreSolarEnergy;
           perform action StroreBatteryPower;
        }
        part propulsionsystem{
            attribute Position;
            perform action ProvideThrust;
            
        }
        part motorsystem{
            
            // Attributes
           attribute speed: Integer; // Speed of the motor in m/s
           attribute altitude: Integer; // Altitude in meters
           attribute power: Integer; // Power in watts

           // Actions/Methods
        perform action speedControl; // Adjust motor speed
        perform action powerDistributionBoard; // Enable or disable power distribution
        perform action monitorTemperature; // Monitor motor temperature
        //perform action emergencyStop; // Stop motor in emergencies
            
        }
        part flightcontrolboard{
             attribute Signal;
             attribute volt;
            perform action stabilizePlatform;
            perform action setNavigationPath;
            //perform action monitorSystemHealth(): string;
           // perform action processSensorData(sensorType: string); 
        }
        part structure{}
        part  navigationcontrol {    }
           
    }
     
    part  payload {
         
         
        perform action ProcessData;
           perform action TransmitData;
       
        
            part CameraSystem{
              attribute resolution;
              attribute storageCapacity;
              perform action scanning;
              perform action recording;
              perform action ImageMosaicking;
              perform action storeData;
                
            }   
        
        
            part SyntheticApertureRadar{
                
                attribute Signal;
                attribute  targetPosition;

                perform action SignalTransmission;
                perform action MovingTargetDectection;
                perform action trakingChange;
                
            }
        
      
            
            part SpectroscopySystem{
                attribute time;
                attribute image;
                attribute scale;
           perform action RealTimeMonitoring;
                
          
            }
        
      
        
    }
    part  GroundStation {
        part communication{
            
            attribute data;
            perform action Reception;
            perform action processData;
            
        }
       
    }
      }
}
```
