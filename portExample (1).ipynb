```sysml
package 'HAPPortSystem' {

  
   part def Data;
    part def Signal;
    

   port def DataOutPort {
        
        out item monitoredData: Data;
        in item processedData : Data;
   }

    port def DataInPort {
       
       out item processedData: Data;
       in item monitoredData : Data;
    }

     port def SendtoGroundData{
         out item sendData :Data;
         in item  receiveData:Data;
     }
    port def ReceivetoGroundData{
         out item receiveData :Data;
         in item  sendData:Data;
     }
     
      port def SignalOutPort {
        out item signalOut: Signal;
        in item signalIn: Signal;
    }
     port def SignalInPort {
         in item signalIn: Signal;
        out item signalOut: Signal;
        
    }
    part def HAP {
        port datainport : DataInPort; 
        port sendtogrounddata: SendtoGroundData;
        port signaloutport : SignalOutPort;
    }
     
   
    part def GroundStation {
        
       port receivetogrounddata: ReceivetoGroundData;
    }
    
    part def MonitorEnv{
        port signalinport : SignalInPort;
        port dataoutport: DataOutPort;
    }
    
    
}
```
