```sysml
package systemStructure {
   
    part def Data;
   part def Signal;

  port def DataPort {
      in item dataIn : Data;
      out item dataOut : Data;
  }

  interface def DataInterface {
      end sourcePort : DataPort;
      end targetPort : DataPort;

      flow sourcePort.dataOut to targetPort.dataIn;
  }

  part def HighAltitudePlatformSystem {
      part payload {
          part EarthObservationSubsystem {
              port dataPort : DataPort;
              attribute resolution;
          }

          part EnvironmentMonitoringSubsystem {
              port dataPort : DataPort;
               attribute coverageArea ;
            
          }

          part SurveillanceSubsystem {
              port dataPort : DataPort;
              attribute detectionRange;
             
          }

          part TrackingSubsystem {
              port dataPort : DataPort;
              attribute trackingAccuracy;
             
          }

          part DisasterMonitoringSubsystem {
              port dataPort : DataPort;
              attribute responseTime;
          }

          part AirQualitySubsystem {
              port dataPort : DataPort;
              attribute pollutantDetectionLimit;
             
          }
          part dataProcessingSubsystem {
              port dataPort : DataPort;
          }


          interface : DataInterface connect
              sourcePort ::> EarthObservationSubsystem.dataPort to
              targetPort ::> dataProcessingSubsystem.dataPort;
      }
  }
}
```
