```sysml
package HAPEnumerationDefinitions {
    
    import ScalarValues::Real;

    enum def HAPPowerState {
        enum solar_day;   
        enum battery_night; 
        enum fuel_cell;    
    }
    
    part def HAPPowerSystem {
        attribute currentState : HAPPowerState;  
    }
    
    part def HAPDaylightOperation specializes HAPPowerSystem {
        attribute redefines currentState = HAPPowerState::solar_day;  
    }
}


```

```sysml
%viz  HAPEnumerationDefinitions  
```

```sysml
package 'HAPClassification' {
    
    import ScalarValues::*;
    import HAPEnumerationDefinitions::*;

    attribute def PowerClassificationLevel {
        attribute code : String;              
        attribute powerState : HAPPowerState;  
    }

    enum def PowerClassificationKind specializes PowerClassificationLevel {
        normal_operation {
            :>> code = "NORM";                      
            :>> powerState = HAPPowerState::solar_day; 
           
        }
        night_transition {
            :>> code = "TWIL";                    
            :>> powerState = HAPPowerState::battery_night;  
        }
        emergency_power {
            :>> code = "EMER";                     
            :>> powerState = HAPPowerState::fuel_cell;     
        }
    }
    }
```
